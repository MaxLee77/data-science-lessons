### 6-VAZIFA. 2010-2020-yillarda AQSH turli manbaalardan o'rtacha qancha elektr toki olganini chiqaring. Natijani kichikdan kattaga qarab tartiblang.

```python
# JAVOBNI SHUYERGA YOZING
avg_elec_2010_2020 = dfus.iloc[25:36, 2:]
avg_elec_2010_2020
```

```python
# @title quyosh vs komir

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['quyosh']
  ys = series['komir']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = avg_elec_2010_2020.sort_values('quyosh', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('quyosh')
_ = plt.ylabel('komir')
```

```python
#@title O'xshash natija (1990-2000 yillar uchun)
```
